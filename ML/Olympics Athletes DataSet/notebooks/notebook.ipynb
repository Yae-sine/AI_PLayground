```python
df.drop_duplicates(inplace= True)

total_medals_by_gender= df.groupby('gender')['total_medals_won'].sum().reset_index()
print(total_medals_by_gender)
```

```python
sns.barplot(data = total_medals_by_gender , x = 'gender' , y = 'total_medals_won' , palette= 'viridis')
```

```python
total_medals_by_year = df.groupby('year')['total_medals_won'].sum().reset_index()
print(total_medals_by_year)


sns.barplot(total_medals_by_year , x = 'year' , y = 'total_medals_won' , palette= 'viridis')
```

```python
numerical_dataset = df.select_dtypes(include=[np.number])

correlation_matrix = numerical_dataset.corr(method = 'pearson')

plt.figure(figsize=(10 , 8 ))
sns.heatmap(correlation_matrix , annot= True , cmap = 'coolwarm', fmt = '.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
total_medals_by_teamOrIndivudal = df.groupby('team_or_individual')['total_medals_won'].sum().reset_index()


sns.barplot(total_medals_by_teamOrIndivudal , x = 'team_or_individual' , y = 'total_medals_won')
print(total_medals_by_teamOrIndivudal)
```

```python
total_medals_by_sport = df.groupby('sport')['total_medals_won'].sum().reset_index()

sns.barplot(total_medals_by_sport , x = 'sport' , y = 'total_medals_won')
print(total_medals_by_sport)
```

```python
total_medals_by_nationality = df.groupby('nationality')['total_medals_won'].sum().reset_index()

sns.barplot(total_medals_by_nationality , x = 'nationality' , y = 'total_medals_won')
print(total_medals_by_nationality)
```

```python
df = df.drop(columns= ['athlete_id' , 'athlete_name' , 'date_of_birth' , 'nationality'])
encoder = OneHotEncoder(sparse_output= False)
X_cat = df[['gender' , 'team_or_individual' , 'games_type' , 'is_record_holder' , 'notes' , 'coach_name' , 'host_city' , 'sport' ,'country_name', 'medal']]


encoded_data = encoder.fit_transform(X_cat)
encoded_df = pd.DataFrame(
    encoded_data ,
    columns= encoder.get_feature_names_out(X_cat.columns) , 
    index = df.index
)

df_encoded = df.drop(columns= X_cat.columns).join(encoded_df)
```
