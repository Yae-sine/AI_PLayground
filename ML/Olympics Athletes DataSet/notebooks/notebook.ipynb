```python
df.drop_duplicates(inplace= True)

total_medals_by_gender= df.groupby('gender')['country_total_medals'].sum().reset_index()
print(total_medals_by_gender)
```

```python
sns.barplot(data = total_medals_by_gender , x = 'gender' , y = 'country_total_medals' , palette= 'viridis')
```

```python
total_medals_by_year = df.groupby('year')['country_total_medals'].sum().reset_index()
print(total_medals_by_year)


sns.barplot(total_medals_by_year , x = 'year' , y = 'country_total_medals' , palette= 'viridis')
```

```python
numerical_dataset = df.select_dtypes(include=[np.number])

correlation_matrix = numerical_dataset.corr(method = 'pearson')

plt.figure(figsize=(10 , 8 ))
sns.heatmap(correlation_matrix , annot= True , cmap = 'coolwarm', fmt = '.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
