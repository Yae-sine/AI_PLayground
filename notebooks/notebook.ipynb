```python
data_2 = data.drop(["Cabin" , "PassengerId" , "Name"] , axis = 1)

data_2["Age"] = data_2["Age"].fillna(data_2["Age"].mean())


imputer = SimpleImputer(strategy="most_frequent")
data_2["Embarked"] = imputer.fit_transform(data_2[["Embarked"]]).ravel()


print(data_2.isnull().sum())



############################################### test 
test_data2 = test_data.drop(["Cabin" , "PassengerId" , "Name"] , axis = 1)
test_data2["Age"] = test_data2["Age"].fillna(test_data2["Age"].mean())
test_data2["Fare"] = test_data2["Fare"].fillna(test_data2["Fare"].mean())

imputer = SimpleImputer(strategy="most_frequent")
test_data2["Embarked"] = imputer.fit_transform(test_data2[["Embarked"]]).ravel()


```

```python
encoder = OneHotEncoder()
encoded_array = encoder.fit_transform(data_2[["Sex" , "Embarked"]]).toarray()

encoded_df = pd.DataFrame(encoded_array)


freq_encoding = data_2['Ticket'].value_counts()

data_2['Ticket_freq'] = data_2['Ticket'].map(freq_encoding)


final_df = pd.concat([data_2 , encoded_df ] , axis=1)
final_df = final_df.drop(["Ticket", "Sex", "Embarked"] , axis = 1)
final_df.columns  = final_df.columns.astype(str)
print(final_df.head())



###########################################


encoded_array_test = encoder.fit_transform(test_data2[["Sex" , "Embarked"]]).toarray()

encoded_test_df = pd.DataFrame(encoded_array_test)


freq_encoding_test = test_data2['Ticket'].value_counts()

test_data2['Ticket_freq'] = test_data2['Ticket'].map(freq_encoding_test)


final_df_test = pd.concat([test_data2 , encoded_test_df ] , axis=1)
final_df_test = final_df_test.drop(["Ticket", "Sex", "Embarked"] , axis = 1)
final_df_test.columns  = final_df_test.columns.astype(str)
```
